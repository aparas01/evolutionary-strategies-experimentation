```python
import numpy as np
import pandas as pd
import os
np.set_printoptions(suppress=True)
pd.set_option('display.max_columns', None)
```

```python
RESULTS_DIR = './es_results'
PROBLEMS_SIZE = 24
```

```python
results_df = pd.DataFrame(columns=['filename', 'mean_error'] + [f'p{i+1}' for i in list(range(PROBLEMS_SIZE))])
for filename in os.listdir(RESULTS_DIR):
    print(f'Found file {filename}')
    results = np.load(os.path.join(RESULTS_DIR, filename))
    mean_error = np.mean(results)
    results_df.loc[len(results_df)] = [filename, mean_error] + list(results)
```

```python
results_df.sort_values(by='mean_error')
```
